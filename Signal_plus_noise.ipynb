```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from scipy.optimize import fsolve
from numpy.linalg import inv
from scipy.stats import norm
```

```python
N=1000
alpha=0.05
z=norm.ppf(1-alpha/2)
```

```python
f_inverse11=[]
f_inverse22=[]
h_inverse11=[]
h_inverse22=[]
mu_MLE=[]
sigma_square_MLE=[]
F_inverse=[]
H_inverse=[]
while len(f_inverse22)<N:
    n=5000
    q=[]
    X=[]
    for i in range(n):
        q.append(0.1*(i+1)%10)
    for j in range(n):
        X.append(np.random.normal(10,(10+q[j])**0.5))

    def ML(paramlist):
        mu_hat,sigma_hat_square=paramlist[0],paramlist[1]
        s1=0
        s2=0
        for i in range(n):
            s1=s1-(X[i]-mu_hat)/(sigma_hat_square+q[i])
            s2=s2+(1/(sigma_hat_square+q[i]))/2-((X[i]-mu_hat)**2/(sigma_hat_square+q[i])**2)/2
        return s1,s2

    s=fsolve(ML,[10,10])
    mu=s[0]
    sigma_square=s[1]
    
    if sigma_square<100 and sigma_square>0:
        mu_MLE.append(mu)
        sigma_square_MLE.append(sigma_square)
        f11=0
        f22=0
        h22=0
        h11=0
        h12=0
        for i in range(n):
            f22=f22+(1/(2*n))/(sigma_square+q[i])**2
            f11=f11+(1/n)/(sigma_square+q[i])
            h11=h11+(1/n)/(sigma_square+q[i])
            h12=h12+(X[i]-mu)/(n*(sigma_square+q[i])**2)
            h22=h22-(1/(2*n))/(sigma_square+q[i])**2+(X[i]-mu)**2/(n*(sigma_square+q[i])**3)
        
        f_inverse11.append(1/f11)
        f_inverse22.append(1/f22)
        h_inverse11.append(h22/(h11*h22-h12**2))
        h_inverse22.append(h11/(h11*h22-h12**2))
        H=[[h11,h12],[h12,h22]]
        F=[[f11,0],[0,f22]]
        F_inverse.append(inv(F))
        H_inverse.append(inv(H))
    else:
        continue
    print(len(f_inverse22))
```

```python
ff=0
hh=0
for i in range(N):
    ff=ff+F_inverse[i]
    hh=hh+H_inverse[i]
```

```python
V=np.cov(mu_MLE,sigma_square_MLE)*n
```

```python
gh11=[]
gf11=[]
gh22=[]
gf22=[]
for i in range(len(f_inverse22)):
    gh11.append(2*norm.cdf(z*(h_inverse11[i]/Vn1)**0.5)-1)
    gf11.append(2*norm.cdf(z*(f_inverse11[i]/Vn1)**0.5)-1)
    gh22.append(2*norm.cdf(z*(h_inverse22[i]/Vn2)**0.5)-1)
    gf22.append(2*norm.cdf(z*(f_inverse22[i]/Vn2)**0.5)-1)
```

```python
dF=[]
dH=[]
for i in range(len(F_inverse)):
    dF.append(np.linalg.norm(F_inverse[i]-V))
    dH.append(np.linalg.norm(H_inverse[i]-V))
```
